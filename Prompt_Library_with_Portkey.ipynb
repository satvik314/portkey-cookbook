## Prompt Library by Portkey
- Store all your prompts at one place!
- Easy to use same prompts across teams.
- Create your prompt library here: https://app.portkey.ai/organisation/prompt_library

```python
# Install Portkey
!pip install -qU portkey-ai
```

```python
# Saving OpenAI API Key

import os
from google.colab import userdata

os.environ['OPENAI_API_KEY'] = userdata.get('PORTKEY_API_KEY')
```

```python
# Retreiving and using prompt using Prompt Library

from portkey_ai import Portkey

client = Portkey(
  api_key=userdata.get('PORTKEY_API_KEY')
)

# Make the prompt creation call with the variables
prompt_completion = client.prompts.completions.create(
  prompt_id="pp-mcq-prompt-03576c",
  variables={"number":"5","topic":"Road Safety"}
)

print(prompt_completion)
```
